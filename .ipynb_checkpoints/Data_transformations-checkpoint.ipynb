```python
import pandas as pd
```

```python
data = pd.read_excel('cleaned_data.xlsx')
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data['pct_change'] = (data['CurrentPrice'] / data['PreviousClose'])-1
```

```python
data['B/M'] = data['CurrentPrice'] / data['BookValue']
```

```python
data['pct_change'].isin([0]).sum(axis=0)
```

```python
data
```

```python
median = data['MarketCap'].median()
```

```python
median
```

```python
cols = data.columns
```

```python
cols
```

```python
large = pd.DataFrame(columns = cols)
```

```python
large_rows = []
small_rows = []
```

```python
list(data.iloc[0])
```

```python
for i in data.index:
    if data['MarketCap'][i] > median:
        large_rows.append(list(data.iloc[i]))
    else:
        small_rows.append(list(data.iloc[i]))
```

```python
large = pd.DataFrame(large_rows, columns=cols)
small = pd.DataFrame(small_rows, columns=cols)
```

```python
large.sort_values('B/M', inplace=True)
small.sort_values('B/M', inplace=True)
```

```python
large
```

```python
x = round(62 * 0.3)
```

```python
big_growth = large[0:x]
big_value = large[-x:]
big_neutral = large[x:-x]
```

```python
small_growth = small[0:x]
small_value = small[-x:]
small_neutral = small[x:-x]
```

```python
rfr = (0.67*4) / 365
nzx = -0.00981
```

```python
# Need excess return of market(market risk premium), SMB, HML, excess return of the stock. & date. 
```

```python
list(small_growth['pct_change'])
```

```python
avg_small = pd.concat([small_growth['pct_change'], small_neutral['pct_change'], small_value['pct_change']]).mean()
```

```python
avg_large = pd.concat([big_growth['pct_change'], big_neutral['pct_change'], big_value['pct_change']]).mean() 
```

```python
SMB = avg_small - avg_large
```

```python
SMB
```

```python
avg_value = pd.concat([big_value['pct_change'], small_value['pct_change']]).mean()
```

```python
avg_growth = pd.concat([big_growth['pct_change'], small_growth['pct_change']]).mean()
```

```python
HML = avg_value - avg_growth
```
